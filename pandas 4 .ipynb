```python
df = pd.DataFrame({'col1' : [1,2,3,4,5],
              'col2' : [45,56,67,34,34],
              'col3': "sudhanshu kumar krish ineuron one".split()
     })
```

```python
df
```

```python
df['col4'] = df['col1']**2
```

```python
df
```

```python
df['col5'] = df['col3'].str[0]
```

```python
df
```

```python
#just like mapper , we have apply fuction for data frame
```

```python
def test(a):
    return a**2
```

```python
test(3)
```

```python
#function applied to df

df['col_0'] = df['col1'].apply(test)
```

```python
df
```

```python
def test1(a):
    if a == 's':
        return 's'
    else:
        return 'z'
```

```python
df['col6'] = df['col5'].apply(test1)
```

```python
df
```

```python
df['col7'] = df['col5'].apply(lambda a: 's' if a == 's' else 'z')
```

```python
df
```
