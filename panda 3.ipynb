```python
df.isnull().sum()    # count of null in each columns     axis=0 (default , rows)  axis=1(columns)
```

```python
df.dropna(axis=1, thresh=800)   # thresh hold - MIN NON-NA values to be present else delete that column.
```

```python
df.dropna(axis=1, thresh=800, how=all)     # how = all/any, all -- it checks if all values are NA then drop  | any -- if any one value is NA then delete that entire row or column
```

```python
df.fillna(value='suraj')   # fill value NA with 'suraj'
```

```python
df.fillna(value=df['Age'].mean())   # fill all NA with avg of age
```

```python
df["Age"].fillna(value=df['Age'].mean(),inplace=True)                     # fill NA in "age" column wil avg age of all.
```
